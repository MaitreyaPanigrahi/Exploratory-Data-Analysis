## Black Friday Dataset EDA 
### Cleaning and preparing the data for Analysis

```python
## Dataset Link : https://www.kaggle.com/datasets/sdolezel/black-friday/data?select=train.csv
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the dataset
df_train = pd.read_csv('train.csv')
```

```python
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_test.head()
```

```python
## Merge both train and test data
df = pd.concat([df_train,df_test])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns='User_ID',inplace =True)
```

```python
df.head()
```

### Data Preprocessing

```python
# Handling the categorical feature Gender
def gender_change(gender):
    if gender == 'M':
        return 1
    else:
        return 0
df['Gender'] = df['Gender'].apply(gender_change)
```

```python
df.head()
```

```python
## Handling categorical Feature Age
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17' : 1,'18-25':2,'26-35':3,
                           '36-45':4,'46-50':5,'51-55':6 ,'55+':7})
df.head()
```

```python
## Handling Categorical feature City_Category
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
df = pd.concat([df,df_city],axis = 1)
```

```python
df.head()
```

```python
df.drop(columns='City_Category',inplace=True)
```

```python
df.head()
```

```python
## Handling Missing Values
df.isnull().sum()
```

```python
## Focus on replacing missing values
## Handling Product_Category_2 missing values
df['Product_Category_2'].value_counts()
```

```python
## Replace the missing value with mode
df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0],inplace=True)
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].value_counts()
```

```python
## Handling Product_Category_3 missing values
df['Product_Category_3'].mode().values[0]
```

```python
## Replace NaN values with mode
df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0],inplace=True)
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
def change_curr_city_years(years):
    if years == '4+':
        return 4
    else:
        return years
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].apply(change_curr_city_years)
```

```python
df.head()
```

```python
df.info()
```

```python
## Convert Stay_In_Current_City_Years as Object to integers
df['Stay_In_Current_City_Years'].unique()
```

```python

df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype('int64')
```

```python
df.info()
```

```python
df[['B','C']] = df[['B','C']].astype('int64')
```
